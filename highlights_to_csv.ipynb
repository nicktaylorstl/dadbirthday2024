```python
import requests

def get_highlight_url(game_id, event_id):
    headers = {
        'Host': 'stats.nba.com',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:72.0) Gecko/20100101 Firefox/72.0',
        'Accept': 'application/json, text/plain, */*',
        'Accept-Language': 'en-US,en;q=0.5',
        'Accept-Encoding': 'gzip, deflate, br',
        'x-nba-stats-origin': 'stats',
        'x-nba-stats-token': 'true',
        'Connection': 'keep-alive',
        'Referer': 'https://stats.nba.com/',
        'Pragma': 'no-cache',
        'Cache-Control': 'no-cache'
    }

    url = 'https://stats.nba.com/stats/videoeventsasset?GameEventID={}&GameID={}'.format(
                event_id, game_id)
    r = requests.get(url, headers=headers)
    json = r.json()
    video_urls = json['resultSets']['Meta']['videoUrls']
    playlist = json['resultSets']['playlist']
    video_event = {'video': video_urls[0]['lurl'], 'desc': playlist[0]['dsc']}
    return (video_event['video'],video_event['desc'])
```

```python
import csv

# Define the file path
file_path = r'D:\Documents\GFA\SportsPython\Basketball\dadbirthday2024\data\heat_play_by_play_2023.csv'
# List to store tuples of (game_id, eventnum)
game_event_list = []

# Open the CSV file
with open(file_path, mode='r') as file:
    # Create a CSV reader object
    csv_reader = csv.reader(file)
    # Skip the header row if exists
    next(csv_reader, None)
    # Iterate through each row in the CSV file
    for row in csv_reader:
        # Assuming game_id is in column 0 and eventnum is in column 1
        game_id = row[0]
        eventnum = row[1]
        # Append a tuple (game_id, eventnum) to the list
        game_event_list.append((game_id, eventnum))

# Print the list of tuples
print(game_event_list)
```

```python


highlights_list = []

# Assuming the file path is 'heat_urls.csv'
csv_file_path = 'data/heat_urls.csv'

count = 1371
# Iterate over game_event_list
with open(csv_file_path, mode='a', newline='') as file:
    writer = csv.writer(file)

    for item in game_event_list[1372:]:
        count +=1
        game_id = item[0]
        eventnum = item[1]
        try:
            url_desc = get_highlight_url(game_id, eventnum)
            url = url_desc[0]
            pk = f"{game_id}{eventnum}"
            desc = url_desc[1]
            row = [pk, game_id, eventnum, url, desc]
            highlights_list.append(row)
            print(f"row number {count}")
            writer.writerow(row)
        except Exception as e:
            # If an error occurs, print the error message and move to the next item
            print(f"Error processing row {count}: {e}")
            continue
```
